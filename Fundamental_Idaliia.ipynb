```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from sec_edgar_downloader import Downloader

print("works")
```

```python
import os  # библиотека для работы с файлами и папками

# Создаем главную папку проекта
PROJECT_PATH = "/content/nvda_agent"

folders = ["data_raw", "data_clean", "reports"]

for f in folders:
    os.makedirs(f"{PROJECT_PATH}/{f}", exist_ok=True)

print("Папки проекта:")
print(os.listdir(PROJECT_PATH))
```

```python
TICKER = "NVDA"
ticker = yf.Ticker(TICKER)

income_raw   = ticker.income_stmt
balance_raw  = ticker.balance_sheet
cashflow_raw = ticker.cashflow

# 4️⃣ сохраняем «как есть» в data_raw (на всякий случай)
income_raw.to_csv (f"{PROJECT_PATH}/data_raw/income_stmt_raw.csv")
balance_raw.to_csv(f"{PROJECT_PATH}/data_raw/balance_sheet_raw.csv")
cashflow_raw.to_csv(f"{PROJECT_PATH}/data_raw/cashflow_raw.csv")

# 5️⃣ функция: взять последние n столбцов (самые старые годы)
def last_n_years(df, n=5):
    total_cols = df.shape[1]               # сколько столбцов всего
    n = min(n, total_cols)                 # если меньше 5 — возьмём сколько есть
    return df.iloc[:, -n:].copy()          # берём САМЫЕ ПОСЛЕДНИЕ n столбцов

# 6️⃣ применяем к трём таблицам
income_5y = last_n_years(income_raw, 5)
balance_5y = last_n_years(balance_raw, 5)
cashflow_5y = last_n_years(cashflow_raw, 5)

# 7️⃣ сохраняем версии «готовые к очистке»
income_5y.to_csv  (f"{PROJECT_PATH}/data_clean/income_stmt_5y.csv")
balance_5y.to_csv (f"{PROJECT_PATH}/data_clean/balance_sheet_5y.csv")
cashflow_5y.to_csv(f"{PROJECT_PATH}/data_clean/cashflow_5y.csv")

# 8️⃣ проверяем размеры таблиц
print("Форма таблиц (строки, столбцы):")
print("Income  :", income_5y.shape)
print("Balance :", balance_5y.shape)
print("Cashflow:", cashflow_5y.shape)
```
